```python
import sys
import os
import matplotlib.pyplot as plt
import numpy as np
sys.path.append('..')
from load_data import load_data
from config import *
```

```python
imgs, msks = load_data(nlst_dir, cases=5)
```

```python
def bbox_3D(img):

    x = np.any(img, axis=(0, 1))
    y = np.any(img, axis=(0, 2))
    z = np.any(img, axis=(1, 2))

    xmin, xmax = np.where(x)[0][[0, -1]]
    ymin, ymax = np.where(y)[0][[0, -1]]
    zmin, zmax = np.where(z)[0][[0, -1]]

    return xmin, xmax, ymin, ymax, zmin, zmax
```

```python
for img, msk in zip(imgs[1:], msks[1:]):
  
    mid_slice = img.shape[0] // 2
    plt.figure(figsize=(10, 8))
    plt_msk = np.ma.masked_where(msk == np.min(msk), msk)    
    plt.imshow(img[mid_slice, :, :], cmap='gray')
    plt.imshow(plt_msk[mid_slice, :, :], alpha=0.15)
    
    xmin, xmax, ymin, ymax, zmin, zmax = bbox_3D(msk)
    
    plt.figure(figsize=(10, 8))
    plt_msk = np.ma.masked_where(msk == np.min(msk), msk)    
    plt.imshow(img[mid_slice, ymin:ymax, xmin:xmax], cmap='gray')
    plt.imshow(plt_msk[mid_slice, ymin:ymax, xmin:xmax], alpha=0.15)
```
